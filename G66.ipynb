# Data Preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
# Separate features and target variable
X = dataset.drop("Cardiovascular_risk(y)", axis=1)
y = dataset["Cardiovascular_risk(y)"]

print ("Shape of original dataset, train:", dataset.shape)
print ("X: shape=", X.shape, "type=", type(X))
print ("y: shape=", y.shape, "type=", type(y))

# Identify categorical and numerical columns
categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()
numerical_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
```

```python
# Encoding categorical variables
# OneHotEncoder for categorical features
categorical_trans = OneHotEncoder(drop="first", sparse_output=False)

# StandardScaler for numerical features
numerical_trans = StandardScaler()

# Apply the LabelEncoder to the target variable
lbl_encoder = LabelEncoder()
y_encoded = lbl_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
print ("full set shape=", X.shape)
print ("train shape=", X_train.shape)
print ("test shape=", X_test.shape)
```

```python
# Separate the numerical and categorical features
X_train_num = X_train[numerical_cols]
X_train_cat = X_train[categorical_cols]

print("X_train_num shape:", X_train_num.shape)
print("X_train_cat shape:", X_train_cat.shape)
```

# Handling Missing Values

```python
print("Number of missing values for all columns:")
print(X_train_num.isnull().sum())

imputer = SimpleImputer(strategy="median")
imputer.fit(X_train_num)
X_train_num_tr = imputer.transform(X_train_num)
```

```python
print("Number of missing values for categorical data:")
print(X_train_cat.isnull().sum())

from sklearn.preprocessing import LabelEncoder
X_train_cat_tr=X_train_cat.apply(LabelEncoder().fit_transform)
print(X_train_cat_tr)
```

```python
scaler = StandardScaler(copy=False)
scaler.fit(X_train_num_tr)       
X_train_num_tr = scaler.transform(X_train_num_tr)

print('\nMean of all columns:\n', X_train_num_tr.mean(axis = 0))
print('\nStd of all columns:\n', X_train_num_tr.std(axis = 0))
```

```python
encoder=LabelEncoder()
y_train_tr = encoder.fit_transform(y_train)
```

```python
# Preprocess the training and testing data
X_train_processed = np.hstack([X_train_num_tr, X_train_cat_tr])

cat_cols = X_test.select_dtypes(include=["object"]).columns.tolist()
num_cols = X_test.select_dtypes(include=["int64", "float64"]).columns.tolist()
X_test_cat = X_test[cat_cols]
X_test_num = X_test[num_cols]

imputer = SimpleImputer(strategy="median")
imputer.fit(X_test_num)
X_test_num_tr = imputer.transform(X_test_num)

scaler = StandardScaler(copy=False)
scaler.fit(X_test_num_tr)       
X_test_num_tr = scaler.transform(X_test_num_tr)

X_test_cat_tr = X_test_cat.apply(LabelEncoder().fit_transform)
encoder = LabelEncoder()
y_test_tr = encoder.fit_transform(y_test)

X_test_processed = np.hstack([X_test_num_tr, X_test_cat_tr])

# Display the processed data shapes
print("X_train_processed shape:", X_train_processed.shape)
print("y_train shape:", y_train.shape)
print("X_test_processed shape:", X_test_processed.shape)
print("y_test_tr shape:", y_test_tr.shape)
```

# Select and Train Model
### Build a Logistic Regression Model, a Random Forest Classification Model, a Support Vector Machine Model

### Confusion Matrix, Accuracy, Precision and Recall

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, confusion_matrix
```

```python
# First algorithm (Logistic Regression)
from sklearn.linear_model import LogisticRegression

# Default Settings
# Initialize the Logistic Regression model
logreg = LogisticRegression(max_iter=1000)

# Train the model
logreg.fit(X_train_processed, y_train_tr)

# Predict on the test set
y_pred_logreg = logreg.predict(X_train_processed)

# Evaluate the model
accuracy_logreg = accuracy_score(y_train_tr, y_pred_logreg)
precision_logreg = precision_score(y_train_tr, y_pred_logreg, average="macro")
recall_logreg = recall_score(y_train_tr, y_pred_logreg, average="macro")
conf_matrix_logreg = confusion_matrix(y_train_tr, y_pred_logreg)
class_report_logreg = classification_report(y_train_tr, y_pred_logreg)

print("Logistic Regression with Default Settings\n")
print(f"Logistic Regression Accuracy: {accuracy_logreg:.4f}")
print(f"Logistic Regression Precision: {precision_logreg:.4f}")
print(f"Logistic Regression Recall: {recall_logreg:.4f} \n")
print("Confusion Matrix:\n", conf_matrix_logreg)
print("\nClassification Report:\n", class_report_logreg)

# Custom Settings
# Initialize the Logistic Regression model
logreg_custom = LogisticRegression(C=0.5, solver="liblinear", penalty="l1", random_state=42)

# Train the model
logreg_custom.fit(X_train_processed, y_train_tr)

# Predict on the test set
y_pred_logreg_custom = logreg_custom.predict(X_train_processed)

# Evaluate the model
accuracy_custom = accuracy_score(y_train_tr, y_pred_logreg_custom)
precision_custom = precision_score(y_train_tr, y_pred_logreg_custom, average="macro")
recall_custom = recall_score(y_train_tr, y_pred_logreg_custom, average="macro")
conf_matrix_custom = confusion_matrix(y_train_tr, y_pred_logreg_custom)
class_report_custom = classification_report(y_train_tr, y_pred_logreg_custom)

print("\nLogistic Regression with Custom Settings\n")
print(f"Logistic Regression Accuracy: {accuracy_custom:.4f}")
print(f"Logistic Regression Precision: {precision_custom:.4f}")
print(f"Logistic Regression Recall: {recall_custom:.4f} \n")
print("Confusion Matrix:\n", conf_matrix_custom)
print("\nClassification Report:\n", class_report_custom)
```

```python
# Second algorithm (Random Forest Classifier)
from sklearn.ensemble import RandomForestClassifier

# Default Settings
# Initialize the Random Forest model
rf = RandomForestClassifier()

# Train the model
rf.fit(X_train_processed, y_train_tr)

# Predict on the test set
y_pred_rf = rf.predict(X_train_processed)

# Evaluate the model
accuracy_rf = accuracy_score(y_train_tr, y_pred_rf)
precision_rf = precision_score(y_train_tr, y_pred_rf, average="macro")
recall_rf = recall_score(y_train_tr, y_pred_rf, average="macro")
conf_matrix_rf = confusion_matrix(y_train_tr, y_pred_rf)
class_report_rf = classification_report(y_train_tr, y_pred_rf)

print("Random Forest Classifier with Default Settings\n")
print(f"Random Forest Classifier Accuracy: {accuracy_rf:.4f}")
print(f"Random Forest Classifier Precision: {precision_rf:.4f}")
print(f"Random Forest Classifier Recall: {recall_rf:.4f} \n")
print("Confusion Matrix:\n", conf_matrix_rf)
print("\nClassification Report:\n", class_report_rf)

# Custom Settings
# Initialize the Random Forest model
rf_custom = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_custom.fit(X_train_processed, y_train_tr)

# Predict on the test set
y_pred_rf_custom = rf_custom.predict(X_train_processed)

# Evaluate the model
accuracy_rf_custom = accuracy_score(y_train_tr, y_pred_rf_custom)
precision_rf_custom = precision_score(y_train_tr, y_pred_rf_custom, average="macro")
recall_rf_custom = recall_score(y_train_tr, y_pred_rf_custom, average="macro")
conf_matrix_rf_custom = confusion_matrix(y_train_tr, y_pred_rf_custom)
class_report_rf_custom = classification_report(y_train_tr, y_pred_rf_custom)

print("Random Forest Classifier with Custom Settings\n")
print(f"Random Forest Classifier Accuracy: {accuracy_rf_custom:.4f}")
print(f"Random Forest Classifier Precision: {precision_rf_custom:.4f}")
print(f"Random Forest Classifier Recall: {recall_rf_custom:.4f} \n")
print("Confusion Matrix:\n", conf_matrix_rf_custom)
print("\nClassification Report:\n", class_report_rf_custom)
```

# Retrieve Average Cross-Validated Prediction Score Values

```python
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import cross_val_predict

y_train_binarized = label_binarize(y_train_tr, classes=[0, 1, 2])
n_classes = y_train_binarized.shape[1]

logreg_model = OneVsRestClassifier(LogisticRegression(C=0.5, solver="liblinear", penalty="l1", random_state=42))
rf_model = OneVsRestClassifier(RandomForestClassifier(n_estimators=100, random_state=42))
svm_model = OneVsRestClassifier(SVC(kernel="linear", random_state=42, probability=True))
xgb_model = OneVsRestClassifier(XGBClassifier(n_estimators=200, learning_rate=0.1, max_depth=6, subsample=0.8, colsample_bytree=0.8, 
                           objective="binary:logistic", eval_metric="mlogloss", random_state=42))

y_scores_logreg = cross_val_predict(logreg_model, X_train_processed, y_train_binarized, cv=5, method="predict_proba")
y_scores_rf = cross_val_predict(rf_model, X_train_processed, y_train_binarized, cv=5, method="predict_proba")
y_scores_svm = cross_val_predict(svm_model, X_train_processed, y_train_binarized, cv=5, method="predict_proba")
y_scores_xgb = cross_val_predict(xgb_model, X_train_processed, y_train_binarized, cv=5, method="predict_proba")

print("Logistic Regression Predictions Shape:", y_scores_logreg.shape)
print("Random Forest Classifier Predictions Shape:", y_scores_rf.shape)
print("Support Vector Machine (SVM) Predictions Shape:", y_scores_svm.shape)
print("XGBoost Predictions Shape:", y_scores_xgb.shape)
```

```python
y_scores_logreg = y_scores_logreg[:, 1]
y_scores_rf = y_scores_rf[:, 1]
y_scores_svm = y_scores_svm[:, 1]
y_scores_xgb = y_scores_xgb[:, 1]
```

# ROC Curve & AUC Scores

```python
from sklearn.metrics import roc_curve, auc

n_classes = 3  

# Binarize the labels
y_train_binarized = label_binarize(y_train, classes=[0, 1, 2])

y_scores_logreg = cross_val_predict(logreg_model, X_train_processed, y_train, cv=5, method="predict_proba")
y_scores_rf = cross_val_predict(rf_model, X_train_processed, y_train, cv=5, method="predict_proba")
y_scores_svm = cross_val_predict(svm_model, X_train_processed, y_train, cv=5, method="predict_proba")
y_scores_xgb = cross_val_predict(xgb_model, X_train_processed, y_train, cv=5, method="predict_proba")

# Compute ROC curve and ROC AUC for each class and model
roc_auc = {}

for i in range(n_classes):
    # Logistic Regression
    fpr_logreg, tpr_logreg, _ = roc_curve(y_train_binarized[:, i], y_scores_logreg[:, i])
    roc_auc[f'logreg_{i}'] = auc(fpr_logreg, tpr_logreg)

    # Random Forest
    fpr_rf, tpr_rf, _ = roc_curve(y_train_binarized[:, i], y_scores_rf[:, i])
    roc_auc[f'rf_{i}'] = auc(fpr_rf, tpr_rf)

    # SVM
    fpr_svm, tpr_svm, _ = roc_curve(y_train_binarized[:, i], y_scores_svm[:, i])
    roc_auc[f'svm_{i}'] = auc(fpr_svm, tpr_svm)

    # XGBoost
    fpr_xgb, tpr_xgb, _ = roc_curve(y_train_binarized[:, i], y_scores_xgb[:, i])
    roc_auc[f'xgb_{i}'] = auc(fpr_xgb, tpr_xgb)

# Plot ROC curves for each class and each model
plt.figure(figsize=(10, 7))

for i in range(n_classes):
    plt.plot(fpr_logreg, tpr_logreg, lw=3, label=f'Logistic Regression Class {i} (AUC = {roc_auc[f"logreg_{i}"]:.4f})')
    plt.plot(fpr_rf, tpr_rf, lw=3, label=f'Random Forest Class {i} (AUC = {roc_auc[f"rf_{i}"]:.4f})')
    plt.plot(fpr_svm, tpr_svm, lw=3, label=f'SVM Class {i} (AUC = {roc_auc[f"svm_{i}"]:.4f})')
    plt.plot(fpr_xgb, tpr_xgb, lw=3, label=f'XGBoost Class {i} (AUC = {roc_auc[f"xgb_{i}"]:.4f})')

# Random guess line
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')

plt.title("ROC Curves for Logistic Regression, Random Forest, SVM, and XGBoost (Multi-Class)", fontsize=15)
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.legend(loc="lower right")
plt.show()
```

```python
# First Model - Logistic Regression
# Since Logistic Regression does not have the feature_importances_ attribute. To get the importance of features in logistic regression, 
# we can use the coefficients (coef_) instead. These coefficients represent the importance of the features in determining the decision boundary.
feature_importance = best_logreg.coef_[0] 
attributes = X_train_num.columns.tolist() + list(X_train_cat)

# Sort the features by importance
sorted_importance = sorted(zip(abs(feature_importance), attributes), reverse=True)

# Print out feature importance
for importance, attribute in sorted_importance:
    print(f"{importance} - {attribute}")
```
